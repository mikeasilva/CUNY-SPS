# Augment Data

In order to maximize the number of training images I will augment the dataset.  I will produce the mirror image of the images with bounding boxes and then alter the bounding boxes.  This way I can get twice as many training examples for the same amount of human effort to label the data.

```python
from PIL import Image, ImageOps
from glob import glob
from os import path

def create_mirror_image(image_and_coord_path):
    image_path = image_and_coord_path[0]
    coord_path = image_and_coord_path[1]
    # Give some feedback
    print("Mirroring " + image_path)
    # Create the mirror image file
    new_image_path = image_path.split(".")
    new_image_path[1] = new_image_path[1] + "_mirror"
    new_image_path = ".".join(new_image_path)
    if not path.exists(new_image_path):
        im = Image.open(image_path)
        im_mirror = ImageOps.mirror(im)
        im_mirror.save(new_image_path, quality=100)
        # Create the bounding box coords of the mirrored image
        new_coord_path = coord_path.split(".")
        new_coord_path[1] = new_coord_path[1] + "_mirror"
        new_coord_path = ".".join(new_coord_path)
        f = open(new_coord_path, "w")
        for line in open(coord_path):
            line_parts = line.split(" ")
            line_parts[1] = str(1 - float(line_parts[1]))
            new_line = " ".join(line_parts)
            f.write(new_line)
        f.close()
```

```python
for line in open("./iNaturalist/images/Poison Ivy/classes.txt"):
    the_label = line.strip()
    pattern = "./iNaturalist/images/" + the_label + "/*.txt"
    labels = glob(pattern)
    labels.remove("./iNaturalist/images/" + the_label + "\\classes.txt")
    labels = [x for x in labels if "_mirror" not in x]
    for coords_path in labels:
        record_id = int(coords_path.split("\\")[-1].split(".txt")[0])
        pattern = "./iNaturalist/images/" + the_label + "/*" + str(record_id) + "*"
        files = glob(pattern)
        image_path = [x for x in files if x not in coords_path][0]
        image_and_coord_path = (image_path, coords_path)
        create_mirror_image(image_and_coord_path)
```
